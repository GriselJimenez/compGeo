```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define Dip
dip = np.arange(0.1,40,0.1)*np.pi/180
# Define operator error
eo=np.arange(1,6,1)*np.pi/180
for i in range(eo.size):
    # maximum strike error
    es = np.tan(eo[i])/np.tan(dip)
    # Plot each curve
    plt.plot(dip*(180/np.pi),es*(180/np.pi),'k')

# One example
# dip = 5 deg
dipex = 5 
# maximum operator error = 2 deg
eoex = 2 
# maximum strike error
esex = np.tan(eoex*np.pi/180)/np.tan(dipex*np.pi/180)*180/np.pi;
# Plot example
plt.plot([dipex,dipex,0], [0,esex,esex], 'k--')
plt.plot(dipex,esex,'ko')
# Add labels to eo curves
plt.text(8,7.5,'1')
plt.text(9,13,'2')
plt.text(10,17.5,'3')
plt.text(11,21,'4')
plt.text(12,24,'5')
plt.text(33,9,'$ε_o$(°)')
# Axes limits
plt.axis ([0, 40, 0, 90])
# Label axes
plt.xlabel('dip δ(°)')
plt.ylabel('Maximum strike error $ε_s$(°)')
# Add grid
plt.grid()
```
